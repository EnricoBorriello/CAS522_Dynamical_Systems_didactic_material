```python
from scipy.integrate import odeint
import numpy as np
import matplotlib.pyplot as plt
```

```python
# parameters
contact_rate = 0.003*0.25
recovery_rate = 0.3
```

```python
# initial conditions
# total population size = 501
S0 = 999
I0 = 1
R0 = 0
x0 = [S0,I0,R0]
```

```python
# ODEs
def odes(x,t):
    S = x[0]
    I = x[1]
    R = x[2]
    dSdt = -contact_rate*I*S
    dIdt = contact_rate*I*S -recovery_rate*I
    dRdt = recovery_rate*I
    return [dSdt,dIdt,dRdt]
```

```python
t = np.linspace(0,30,1000)
```

```python
x = odeint(odes,x0,t)
```

```python
S = x[:,0]
I = x[:,1]
R = x[:,2]
```

```python
plt.plot(t,S,label='S')
plt.plot(t,I,label='I')
plt.plot(t,R,label='R')
plt.legend()
plt.xlabel('time')
plt.show()
```
