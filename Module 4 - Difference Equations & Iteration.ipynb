```python
import matplotlib.pyplot as plt
```

```python
def logistic_map(r,x):
    return r*x*(1-x)
```

```python
r = 1.5
x0 = .2
T = 20

x = [x0]
for t in range(T):
    x.append(logistic_map(r,x[-1]))
```

```python
plt.plot(range(T+1),x)
plt.plot(range(T+1),x,'ko')
plt.xlabel('t',size=14)
plt.ylabel('x',size=14);
```
